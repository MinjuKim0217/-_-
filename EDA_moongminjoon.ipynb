## 해외 ETF 사용자의 연령대 분포 비교

```python
df['ETF_TRADING_YN_NEW'] = df[['DMETF_FST_BUY_YM', 'OVETF_FST_BUY_YM']].apply(lambda row: int(row.notna().any()), axis=1)

# Calculate the number of people trading ETFs in each age group
age_etf_counts = df.groupby('AGE_TCD')['ETF_TRADING_YN_NEW'].sum()


age_total_counts = df['AGE_TCD'].value_counts().sort_index()


# Calculate the percentage of people in each age group who are trading ETFs
age_all_etf_percentage = (age_etf_counts / age_total_counts * 100).fillna(0)

age_all_etf_percentage
```

```python
# Import the necessary library
import matplotlib.pyplot as plt

# Color palette
color_palette = ['#F0B26B', '#CB6015', '#AD624E', '#84888B', '#8DC8E8', '#0086B8', '#00A9CE', '#7E9FC3', '#7E9FC3']

# Create the bar graph
fig, ax = plt.subplots()
bars = plt.bar(age_all_etf_percentage.index, age_all_etf_percentage, color=color_palette[:len(age_all_etf_percentage)], width=5)

# Add labels and title
plt.xlabel('연령대')
plt.ylabel('ETF 거래 비율 (%)')
plt.title('연령대별 국내+해외 ETF 거래 비율')

# Annotate with the actual values
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, f"{yval:.1f}%", va='bottom', ha='center')  # va: vertical alignment, ha: horizontal alignment

# Show the plot
plt.show()
```

```python
# ETF거래 여부 칼럼 생성
df['ETF_TRADING_YN'] = df['OVETF_FST_BUY_YM'].notna().astype(int)
#연령대별 ETF 거래 횟수 
age_etf_counts = df.groupby('AGE_TCD')['ETF_TRADING_YN'].sum()
print(age_etf_counts)

# total_etf_traders = age_etf_counts.sum()
# print(total_etf_traders)


# 연령대별 전체 인원 수 계산
age_total_counts = df['AGE_TCD'].value_counts().sort_index()
print(age_total_counts)

# 각 연령대에서 ETF를 거래한 인원의 비율을 계산
age_oversea_etf_percentage = (age_etf_counts / age_total_counts * 100).fillna(0)

print(age_oversea_etf_percentage)
```

```python
import matplotlib.pyplot as plt

# The given color palette
color_palette = ['#F0B26B', '#CB6015', '#AD624E', '#84888B', '#8DC8E8', '#0086B8', '#00A9CE', '#7E9FC3', '#7E9FC3']

# Plotting the bar graph
fig, ax = plt.subplots()
bars = plt.bar(age_oversea_etf_percentage.index, age_oversea_etf_percentage, color=color_palette[:len(age_oversea_etf_percentage)], width=5)

# Add labels and title
plt.xlabel('연령대')
plt.ylabel('ETF 거래 비율')
plt.title('연령대별 해외 ETF 거래 비율 (%)')

# Annotate with the actual values
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width(), yval, int(yval), va='bottom')  # va: vertical alignment

plt.show()
```

```python
# Create indicator columns for Domestic, Overseas, and both ETF trading
df['DMETF_TRADING_YN'] = df['DMETF_FST_BUY_YM'].notna().astype(int)
df['OVETF_TRADING_YN'] = df['OVETF_FST_BUY_YM'].notna().astype(int)
df['BOTH_ETF_TRADING_YN'] = df[['DMETF_FST_BUY_YM', 'OVETF_FST_BUY_YM']].apply(lambda row: int(row.notna().all()), axis=1)

# Calculate the number of people trading ETFs in each category by age group
age_dm_counts = df.groupby('AGE_TCD')['DMETF_TRADING_YN'].sum()
age_ov_counts = df.groupby('AGE_TCD')['OVETF_TRADING_YN'].sum()
age_both_counts = df.groupby('AGE_TCD')['BOTH_ETF_TRADING_YN'].sum()

# Calculate the percentage of people in each age group who are trading in each category
age_dm_percentage = (age_dm_counts / age_total_counts * 100).fillna(0)
age_ov_percentage = (age_ov_counts / age_total_counts * 100).fillna(0)
age_both_percentage = (age_both_counts / age_total_counts * 100).fillna(0)
```

```python

# Plotting
fig, ax = plt.subplots(figsize=(15, 6))  # Adjust figsize for a wider plot

# Creating bar positions
x = np.arange(len(age_total_counts.index))

# Width of a bar
width = 0.2

# Plotting each category
bars1 = plt.bar(x - width, age_dm_percentage, width, label='국내 ETF', color=color_palette[0])
bars2 = plt.bar(x, age_ov_percentage, width, label='해외 ETF', color=color_palette[1])
bars3 = plt.bar(x + width, age_both_percentage, width, label='모두', color=color_palette[2])

# Adding labels and title
plt.xlabel('연령대')
plt.ylabel('ETF 거래 비율 (%)')
plt.title('연령대별 국내, 해외, 국내+해외 ETF 거래 비율')

plt.xticks(x, age_total_counts.index) 
# Adding legend
plt.legend()

# Annotate with the actual values
for bars in [bars1, bars2, bars3]:
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2.0, yval, f"{yval:.1f}%", va='bottom', ha='center')

# Show the plot
plt.show()





```

```python

# Initialize an empty DataFrame to hold the summed metrics
summed_df = pd.DataFrame(index=df['AGE_TCD'].unique(), columns=['BUY_CNT', 'SEL_CNT', 'BUY_ITM_CNT', 'SEL_ITM_CNT', 'BUY_AMT', 'SEL_AMT', 'STK_IN', 'STK_OUT', 'MNY_IN', 'MNY_OUT'])

# Loop over each metric and sum across periods (M1, M2, M3, etc.)
for metric in ['BUY_CNT', 'SEL_CNT', 'BUY_ITM_CNT', 'SEL_ITM_CNT', 'BUY_AMT', 'SEL_AMT', 'STK_IN', 'STK_OUT', 'MNY_IN', 'MNY_OUT']:
    matching_columns = [col for col in df.columns if metric in col]  # Find all columns that contain the metric
    summed_df[metric] = df.groupby('AGE_TCD')[matching_columns].sum().sum(axis=1)  # Sum across all matching columns for each age group

# Now, summed_df contains the summed metrics for each age group.
```
